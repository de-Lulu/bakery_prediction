```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load prepared data
df = pd.read_csv('/workspaces/bakery_prediction/0_DataPreparation/merged_data.csv')
```

```python
# Filter for training data only
plot_data = df[df['is_train'] == True].copy()
```

```python
# Map weekday numbers to names
weekday_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
plot_data['WeekdayName'] = plot_data['Weekday'].map(weekday_map)
order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
# Plot
plt.figure(figsize=(10, 6))
# sns.barplot shows the point estimate (mean) and confidence interval by default
sns.barplot(x='WeekdayName', y='Umsatz', data=plot_data, order=order, capsize=.1)
plt.title('Average Sales per Weekday with Confidence Intervals')
plt.xlabel('Weekday')
plt.ylabel('Average Sales (Umsatz)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('sales_per_weekday.png')
```

```python
plt.figure(figsize=(8, 5))
weekday_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
plot_df = df[df['is_train']].copy()
plot_df['WeekdayName'] = plot_df['Weekday'].map(weekday_map)
order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
sns.barplot(x='WeekdayName', y='Umsatz', data=plot_df, order=order, capsize=.1)
plt.title('Sales by Weekday (with 95% CI)')
plt.savefig('plot_weekday.png')
```

```python
# Plot 2: Month
plt.figure(figsize=(8, 5))
sns.barplot(x='Month', y='Umsatz', data=plot_df, capsize=.1)
plt.title('Sales by Month (with 95% CI)')
plt.savefig('plot_month.png')
```
